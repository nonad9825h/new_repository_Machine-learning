### Алгоритм

```python
def parsing_json(url):
    data = {
            "title": [],
            "namecompany": [],
            "description": [],
            "rating": [],
            "field": [],
            "date": [],
            "textpub": [],
            "views": [],
            }
    
    with open(url, 'r', encoding='utf8') as file:
        data_test = json.load(file)

        data_mini = {
            "title": [],
            "namecompany": [],
            "description": [],
            "rating": [],
            "field": [],
            "date": [],
            "textpub": [],
            "views": [],
            }

        
        data_mini['namecompany'] = Path(url).stem
        
        if data_test['info'] is not None:
            info_content = data_test['info']

            field = ''
            for content in info_content['industries']:
                field += content + ', '
            
            data_mini['description'] = info_content['about']
            data_mini['field'] =  field


        
        else:
            data_mini['title'] = 'None'
            data_mini['description'] = 'None'
            data_mini['field'] = 'None'
        
        for content_all in data_test['refs']:
            if content_all is None:
                data_mini['textpub'] = "-"
                data_mini['date'] = "0"
                data_mini['rating'] = "0"
                data_mini['views'] = "0"
            else:
                data_mini['textpub'] = content_all[0]
                data_mini['date'] = content_all[1]['day'] + " " + content_all[1]['month'] + " " + content_all[1]['time']
                data_mini['rating'] = content_all[2]['rate']
                data_mini['views'] = content_all[2]['views']

            # print(data_mini)
            for labels in data:
                data[labels].append(data_mini[labels]) 

    return data  
```

```python
%%time
df = pd.DataFrame(data=result_list)

for link in all_json:
    data = parsing_json(link)
    dataset = pd.DataFrame(data=data)
    df = pd.concat([df, dataset])   

df = df.reset_index(drop=True)
```
